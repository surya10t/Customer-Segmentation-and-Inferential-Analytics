## Full Store Data

```python
s1 = pd.read_csv('Full_factiss_merged_dr_gr.csv')
```

```python
s1.head()
```

```python
s1.shape
```

```python
s1['SY'] = s1['SY'].apply(lambda x: '{0:0>2}'.format(x))
```

```python
s1['GE'] = s1['GE'].apply(lambda x: '{0:0>2}'.format(x))
```

```python
s1['VEND'] = s1['VEND'].apply(lambda x: '{0:0>5}'.format(x))
```

```python
s1['ITEM'] = s1['ITEM'].apply(lambda x: '{0:0>5}'.format(x))
```

```python
s1.shape
```

```python
s1['UPC'] = s1['SY'].map(str) + "-" + s1['GE'].map(str) + "-" + s1['VEND'].map(str) + "-" + s1['ITEM'].map(str)
```

```python
s1.head()
```

```python
s2 = pd.read_csv('prod_tissue_V2.csv')
```

```python
s2.head()
```

```python
s3 = pd.merge(s1, s2, on='UPC', how='left')
```

```python
s3.shape
```

```python
s3.to_csv('Full_gr_dr_prod_merged.csv')
```
